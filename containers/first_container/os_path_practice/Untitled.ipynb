```python
# 1. This creates the app.py file automatically
app_code = """import os

print(f"📂 Current Working Directory: {os.getcwd()}")

log_dir = "logs"
filename = "final_misfire_report.txt"

if not os.path.exists(log_dir):
    os.makedirs(log_dir)

target_filepath = os.path.abspath(os.path.join(log_dir, filename))

print("📍 Exact Hard Drive Location:")
print(target_filepath)

with open(target_filepath, "w") as f:
    f.write("🚨 MISFIRE DETECTED: Cylinder 4 Misfire (P0304)")
"""

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)
print("✅ Success! 'app.py' has been automatically created for you.")


# 2. This creates the Dockerfile automatically
dockerfile_code = """FROM python:3.11-slim
WORKDIR /usr/src/myapp
COPY app.py .
CMD ["python", "app.py"]
"""

with open("Dockerfile", "w", encoding="utf-8") as f:
    f.write(dockerfile_code)
print("✅ Success! 'Dockerfile' has been automatically created for you.")
```

```python
import os

# Let's see where the script is executing from inside the environment
print(f"📂 Current Working Directory: {os.getcwd()}")

# 1. Define the relative directory and filename
log_dir = "logs"
filename = "final_misfire_report.txt"

# 2. Automatically ensure the folder structure exists
if not os.path.exists(log_dir):
    os.makedirs(log_dir)

# 3. Use os.path to calculate the absolute path dynamically
target_filepath = os.path.abspath(os.path.join(log_dir, filename))

print("📍 Exact Hard Drive Location:")
print(target_filepath)

# 4. Write the diagnostic file safely
with open(target_filepath, "w") as f:
    f.write("🚨 MISFIRE DETECTED: Cylinder 4 Misfire (P0304)")
```
