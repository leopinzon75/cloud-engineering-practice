```````python
# 1. MANUFACTURE THE PARTS LIST (requirements.txt)
requirements_content = """boto3==1.34.0
moto[mock_aws]==5.0.0
"""

with open("requirements.txt", "w") as f:
    f.write(requirements_content)
print("📝 Generated: requirements.txt")


# 2. MANUFACTURE THE FACTORY BLUEPRINT (Dockerfile)
dockerfile_content = """FROM python:3.11-slim
WORKDIR /app
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt
COPY cloud_pipeline.py .
COPY baseline_test.txt .
CMD ["python", "cloud_pipeline.py"]
"""

with open("Dockerfile", "w") as f:
    f.write(dockerfile_content)
print("🐳 Generated: Dockerfile")


# 3. MANUFACTURE THE CORE APPLICATION (cloud_pipeline.py)
pipeline_content = """import boto3
from moto import mock_aws

@mock_aws
def cloud_step_four():
    s3 = boto3.resource("s3", region_name="us-east-1")
    s3.create_bucket(Bucket="step-by-step-bucket")
    
    with open("baseline_test.txt", "r") as file:
        local_data = file.read()
    s3.Object("step-by-step-bucket", "cloud_misfire_copy.txt").put(Body=local_data)
    
    cloud_object = s3.Object("step-by-step-bucket", "cloud_misfire_copy.txt")
    cloud_data = cloud_object.get()['Body'].read().decode('utf-8')
    
    print("📥 Data pulled back down from S3 Bucket:")
    print(cloud_data)

if __name__ == "__main__":
    cloud_step_four()
"""

with open("cloud_pipeline.py", "w") as f:
    f.write(pipeline_content)
print("🐍 Generated: cloud_pipeline.py")

print("\n🚀 Setup complete! Your entire Docker workspace has been automated.")
```````

```````python
import boto3
from moto import mock_aws

@mock_aws
def cloud_step_four():
    s3 = boto3.resource("s3", region_name="us-east-1")
    s3.create_bucket(Bucket="step-by-step-bucket")
    
    # 1. READ LOCAL FILE
    with open("baseline_test.txt", "r") as file:
        local_data = file.read()
    
    # ──► INSPECTION 1: Local data type before upload
    print("🔍 1. Local data material type:")
    print(type(local_data))  # This will be <class 'str'>
    print("-" * 40)

    # Upload to Cloud
    s3.Object("step-by-step-bucket", "cloud_misfire_copy.txt").put(Body=local_data)
    
    # 2. PULL DOWN FROM CLOUD (Raw Stream)
    cloud_object = s3.Object("step-by-step-bucket", "cloud_misfire_copy.txt")
    
    # Let's stop right after reading the raw body stream
    raw_bytes_payload = cloud_object.get()['Body'].read()
    
    # ──► INSPECTION 2: Raw data material straight out of the bucket
    print("🔍 2. Raw cloud data material type (Before Decoding):")
    print(type(raw_bytes_payload))  # This will be <class 'bytes'>
    print(f"🖨️ Raw look: {raw_bytes_payload}")
    print("-" * 40)

    # 3. DECODE THE PAYLOAD
    decoded_string = raw_bytes_payload.decode('utf-8')
    
    # ──► INSPECTION 3: Data material after decoding
    print("🔍 3. Final cloud data material type (After Decoding):")
    print(type(decoded_string))  # This will be <class 'str'>
    print(f"🖨️ Clean look: {decoded_string}")

# if __name__ == "__main__":
#     cloud_step_four()
cloud_step_four()
```````
