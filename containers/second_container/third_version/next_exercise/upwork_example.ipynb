```python
#!/usr/bin/env python3
"""
SISTEMA DE PROCESAMIENTO DE DIAGNÓSTICOS DE FLOTA AUTOMATIZADO
Cliente: Upwork Enterprise Case Study
Desarrollador: Cloud Engineering Specialist- cloud enterprise engineering

Este script automatiza la lectura de logs crudos de vehículos, limpia 
las anomalías y escribe un reporte estandarizado en una ruta dinámica.
Apto para ejecución local o dentro de contenedores Docker.
"""

import os
import sys
from datetime import datetime

# ==========================================
# GESTIÓN DINÁMICA DE RUTAS (Fase 1 y 2 de tu Blueprint)
# ==========================================
# Determinamos la raíz de donde se ejecuta el script para evitar rutas fijas tipo 'C:/Users/...'
BASE_DIR = os.path.dirname(os.path.abspath(__file__))

# Definimos las carpetas de trabajo de forma relativa y segura
INPUT_DIR = os.path.join(BASE_DIR, "data_input")
OUTPUT_DIR = os.path.join(BASE_DIR, "logs")
Cloud_engineering_practice/containers/second_container/one/logs

def inicializar_entorno():
    """Garantiza que la infraestructura de carpetas exista antes de procesar."""
    try:
        os.makedirs(INPUT_DIR, exist_ok=True)
        os.makedirs(OUTPUT_DIR, exist_ok=True)
        print(f"📡 [ENTORNO] Infraestructura lista.")
        print(f"   -> Entrada: {INPUT_DIR}")
        print(f"   -> Salida:  {OUTPUT_DIR}")
    except Exception as e:
        print(f"❌ [ERROR] No se pudieron crear las carpetas del sistema: {e}")
        sys.exit(1)


# ==========================================
# LÓGICA DE AUTOMATIZACIÓN Y PROCESAMIENTO
# ==========================================
def procesar_diagnosticos_flota():
    """Lee las métricas de la flota y genera un reporte limpio consolidado."""
    print("⚡ [PROCESO] Iniciando escaneo de datos operativos...")
    
    # Simulamos los datos crudos que el cliente recibe de sus vehículos
    datos_crudos_vehiculos = [
        {"id_unidad": "TRUCK-01", "error_code": "P0300", "estado": "Critico"},
        {"id_unidad": "VAN-04", "error_code": "None", "estado": "Operativo"},
        {"id_unidad": "TRUCK-02", "error_code": "P0171", "estado": "Revision"},
    ]
    
    # Nombre del archivo de salida con estampa de tiempo
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    archivo_salida = os.path.join(OUTPUT_DIR, f"reporte_limpio_{timestamp}.txt")
    
    try:
        print(f"✍️ [ESCRITURA] Escribiendo reporte optimizado en: {archivo_salida}")
        
        with open(archivo_salida, "w", encoding="utf-8") as file:
            file.write("==================================================\n")
            file.write(f"REPORTE CONSOLIDADO DE REVISIÓN DE FLOTA - {datetime.now()}\n")
            file.write("==================================================\n\n")
            
            unidades_procesadas = 0
            for vehiculo in datos_crudos_vehiculos:
                # Filtrado de datos: Al cliente solo le importan unidades con fallas reales
                if vehiculo["error_code"] != "None":
                    linea = f"[ALERTA] Unidad: {vehiculo['id_unidad']} | Código: {vehiculo['error_code']} | Status: {vehiculo['estado']}\n"
                    file.write(linea)
                    unidades_procesadas += 1
            
            file.write(f"\n[RESUMEN] Procesamiento completado. Unidades críticas detectadas: {unidades_procesadas}\n")
            
        print("✅ [ÉXITO] El reporte automatizado se ha generado correctamente.")
        
    except IOError as e:
        print(f"❌ [ERROR] Fallo de entrada/salida al escribir el reporte: {e}")
    except Exception as e:
        print(f"❌ [ERROR] Error inesperado en el procesamiento: {e}")


# ==========================================
# PUNTO DE ENTRADA PRINCIPAL
# ==========================================
if __name__ == "__main__":
    print("🚀 --- INICIANDO PIPELINE DE AUTOMATIZACIÓN DE INFRAESTRUCTURA ---")
    inicializar_entorno()
    procesar_diagnosticos_flota()
    print("🏁 --- PIPELINE FINALIZADO CON ÉXITO ---") 
```

```python
#!/usr/bin/env python3
"""
SISTEMA DE PROCESAMIENTO DE DIAGNÓSTICOS DE FLOTA AUTOMATIZADO
Cliente: Upwork Enterprise Case Study
Desarrollador: Cloud Engineering Specialist

Versión Moderna: Optimizado con Pathlib para compatibilidad total.
"""

import sys
from datetime import datetime
from pathlib import Path  # <-- La librería moderna orientada a objetos

# ==========================================
# GESTIÓN DINÁMICA DE RUTAS CON PATHLIB
# ==========================================
# Path.cwd() funciona perfectamente en Jupyter Notebook y en servidores de producción
BASE_DIR = Path.cwd()

# Usamos el operador "/" que Pathlib sobrecarga para unir rutas de forma elegante
INPUT_DIR = BASE_DIR / "data_input"
OUTPUT_DIR = BASE_DIR / "logs"


def inicializar_entorno():
    """Garantiza que la infraestructura de carpetas exista antes de procesar."""
    try:
        # Pathlib maneja la creación de carpetas directamente desde el objeto de la ruta
        INPUT_DIR.mkdir(parents=True, exist_ok=True)
        OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
        print(f"📡 [ENTORNO] Infraestructura lista con Pathlib.")
        print(f"   -> Entrada: {INPUT_DIR}")
        print(f"   -> Salida:  {OUTPUT_DIR}")
    except Exception as e:
        print(f"❌ [ERROR] No se pudieron crear las carpetas del sistema: {e}")
        sys.exit(1)


# ==========================================
# LÓGICA DE AUTOMATIZACIÓN Y PROCESAMIENTO
# ==========================================
def procesar_diagnosticos_flota():
    """Lee las métricas de la flota y genera un reporte limpio consolidado."""
    print("⚡ [PROCESO] Iniciando escaneo de datos operativos...")
    
    datos_crudos_vehiculos = [
        {"id_unidad": "TRUCK-01", "error_code": "P0300", "estado": "Critico"},
        {"id_unidad": "VAN-04", "error_code": "None", "estado": "Operativo"},
        {"id_unidad": "TRUCK-02", "error_code": "P0171", "estado": "Revision"},
    ]
    
    # Nombre del archivo usando interpolación de cadenas de Pathlib
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    archivo_salida = OUTPUT_DIR / f"reporte_limpio_{timestamp}.txt"
    
    try:
        print(f"✍️ [ESCRITURA] Escribiendo reporte en: {archivo_salida}")
        
        # Abrimos el archivo pasándole directamente el objeto Path
        with open(archivo_salida, "w", encoding="utf-8") as file:
            file.write("==================================================\n")
            file.write(f"REPORTE CONSOLIDADO DE REVISIÓN DE FLOTA - {datetime.now()}\n")
            file.write("==================================================\n\n")
            
            unidades_procesadas = 0
            for vehiculo in datos_crudos_vehiculos:
                if vehiculo["error_code"] != "None":
                    linea = f"[ALERTA] Unidad: {vehiculo['id_unidad']} | Código: {vehiculo['error_code']} | Status: {vehiculo['estado']}\n"
                    file.write(linea)
                    unidades_procesadas += 1
            
            file.write(f"\n[RESUMEN] Procesamiento completado. Unidades críticas: {unidades_processed if 'unidades_processed' in locals() else unidades_procesadas}\n")
            
        print("✅ [ÉXITO] El reporte automatizado se ha generado correctamente.")
        
    except IOError as e:
        print(f"❌ [ERROR] Fallo de entrada/salida al escribir el reporte: {e}")
    except Exception as e:
        print(f"❌ [ERROR] Error inesperado en el procesamiento: {e}")


# ==========================================
# PUNTO DE ENTRADA PRINCIPAL
# ==========================================
if __name__ == "__main__":
    print("🚀 --- INICIANDO PIPELINE DE AUTOMATIZACIÓN (MODO PATHLIB) ---")
    inicializar_entorno()
    procesar_diagnosticos_flota()
    print("🏁 --- PIPELINE FINALIZADO CON ÉXITO ---")
```
