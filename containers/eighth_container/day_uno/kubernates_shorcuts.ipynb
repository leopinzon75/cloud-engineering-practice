Para que sea fácil de escanear, los acomodé por la tarea que vayas a realizar:


1. El Ciclo de Vida (Construir y Arrancar)
Aquí es donde preparas el motor y lo pones a marchar.
Acción                         En Docker (Local)                             En Kubernetes (Clúster)

Construir la imagen            docker build -t mi-app:v1 .                   No aplica (usa la imagen ya construida)

Subir la imagen al almacén     docker push mi-usuario/mi-app:v1              No aplica

Arrancar la aplicación         docker run -d --name mi-contenedor mi-app:v1  kubectl apply -f mi-plano.yaml


Apagar / Borrar todo           docker rm -f mi-contenedor                    kubectl delete -f mi-plano.yaml



2. Inspección e Inventario (¿Qué hay corriendo?)
 
Para revisar el taller y ver qué herramientas o contenedores están activos.

Acción                          En Docker (Local)                             En Kubernetes (Clúster)

Listar lo que está corriendo    docker ps                                     kubectl get pods

Listar TODO (corriendo o apagado)docker ps -a                                 kubectl get pods -A (en todos los Namespaces)

Ver la ficha técnica / Radiografía docker inspect mi-contenedor               kubectl describe pod mi-pod

Listar otros objetos             docker image ls o docker network ls       kubectl get services o kubectl get namespaces



3. Diagnóstico y Reparación (¿Qué está pasando adentro?)
 
Tus herramientas de diagnóstico principales cuando algo no se comporta como debería.

Acción                           En Docker (Local)                             En Kubernetes (Clúster)

Escuchar el código               (Logs) docker logs mi-contenedor              kubectl logs mi-pod

Escuchar logs en tiempo real      docker logs -f mi-contenedor                 kubectl logs -f mi-pod

Meterse al contenedor (Shell)    docker exec -it mi-contenedor sh              kubectl exec -it mi-pod -- sh


💡 Un tip de productividad para Kubernetes: Escribir kubectl a cada rato cansa los dedos. En tu terminal puedes configurar un alias para que solo con presionar la letra k haga lo mismo (ej. k get pods).
