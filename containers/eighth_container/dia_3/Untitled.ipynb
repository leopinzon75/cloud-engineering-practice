```python
%%writefile validador.py
import os
import time
import sys

def generar_reporte():
    app_env = os.getenv("APP_ENV", "Desarrollo-Local")
    ruta_compartida = "/data"
    archivo_reporte = os.path.join(ruta_compartida, "reporte_salud.txt")
    
    print(f"=== INICIANDO VALIDACIÓN EN ENTORNO: {app_env} ===")
    print("Verificando sistemas...")
    time.sleep(1)
    
    # Validamos que la ranura de almacenamiento exista en el contenedor
    if not os.path.exists(ruta_compartida):
        print(f"❌ Error: La ruta {ruta_compartida} no está montada.")
        sys.exit(1)
        
    # Escribimos el reporte en el volumen compartido
    with open(archivo_reporte, "w") as f:
        f.write(f"Reporte de salud para: {app_env}\n")
        f.write("Estado: Sistema en perfecto equilibrio (X).\n")
        
    print(f"✅ Reporte escrito con éxito en: {archivo_reporte}")
    print("=== TRABAJO BATCH FINALIZADO ===")

if __name__ == "__main__":
    generar_reporte()
```

```python
%%bash
kubectl create configmap validador-script-config --from-file=validador.py
```

```python
%%writefile job_validador.yaml
apiVersion: batch/v1
kind: Job
metadata:
  name: job-validador-batch
  labels:
    component: architect-python-batch
    layer: exercise-2
spec:
  backoffLimit: 1
  template:
    metadata:
      labels:
        component: architect-python-batch
        layer: exercise-2
    spec:
      restartPolicy: Never
      containers:
      - name: python-worker
        image: python:3.10-slim
        # Le decimos al contenedor que ejecute el archivo que va a aparecer mágicamente en /app
        command: ["python", "/app/validador.py"]
        
        env:
        - name: APP_ENV
          value: "Produccion-Simulada"
          
        volumeMounts:
        # Montaje 1: La tarjeta con nuestro código de Python
        - name: espacio-codigo
          mountPath: /app
        # Montaje 2: La carpeta temporal donde el script escribirá el reporte
        - name: espacio-datos
          mountPath: /data
          
      volumes:
      # Definimos el volumen que apunta al ConfigMap (la tarjeta)
      - name: espacio-codigo
        configMap:
          name: validador-script-config
      # Definimos el volumen temporal para los datos de salida
      - name: espacio-datos
        emptyDir: {}
```

```python
%%bash
kubectl apply -f job_validador.yaml
```

```python
%%bash
kubectl get jobs,pods -l component=architect-python-batch
```

```python
%%bash
kubectl logs job/job-validador-batch
```
