```python
from fastapi import FastAPI
import uvicorn
import asyncio

app = FastAPI()

@app.get("/")
def read_root():
    return {"Hello": "Youtube"}

# Configuración del servidor de Uvicorn
config = uvicorn.Config(app=app, host="127.0.0.1", port=8000, loop="asyncio")
server = uvicorn.Server(config)

# En lugar de uvicorn.run(), le decimos al bucle de Jupyter que monte el servidor
if __name__ == "__main__":
    loop = asyncio.get_event_loop()
    if loop.is_running():
        # Usamos el bucle existente de Jupyter para crear la tarea del servidor
        loop.create_task(server.serve())
    else:
        # Por si acaso el bucle no estuviera corriendo (fuera de Jupyter)
        loop.run_until_complete(server.serve())
```

```python
%%writefile service.yaml
apiVersion: v1
kind: Service
metadata:
  name: fastapi-service   # El nombre oficial de tu conmutador de red
spec:
  type: NodePort          # Expone el puerto hacia afuera de tu computadora local
  selector:
    app: mi-api           # <-- Crucial: El radar que amarra este servicio a tus 3 réplicas
  ports:
    - protocol: TCP
      port: 80            # El puerto por donde se hablará dentro del clúster
      targetPort: 8000    # El puerto interno de tu contenedor (donde corre Uvicorn)
      nodePort: 30080     # El puerto fijo de tu computadora (entre 30000 y 32767) para entrar desde el navegador
```
